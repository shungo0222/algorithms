# Nodeクラスを利用する方法
---

1. queueに起点を追加する
2. queueがなくなるまで次の処理を繰り返す  
 1. queueからnodeを一つ取り出す。  
    popで取り出すとDFSになって、popleftで取り出すとBFSになる。
 2. 取り出したnodeの隣接nodeをnearsに代入する。
 3. nearsから隣接nodeを一つずつ取り出し、探索済みかどうかを調べる。
 4. 未探索のnodeをqueueにappendする。
 5. queueがなくなる前に目的のnodeが見つかれば探索終了。

```python
class Node(object):
    """
    Class of node

    Parameters
    ----------
    index : int
        the location of node
    nears : list
        the list of adjacent nodes
    visit : bool
        the sign indicating visited or not
        True - visited
        False - not visited
    """

    def __init__(self, index):
        self.index = index
        self.nears = []
        self.visit = False
    
    def __repr__(self):
        return f'Node {{index: {self.index}, nears: {self.nears}, visit: {self.visit}}}'
```
